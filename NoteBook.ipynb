# 3. Learining Algorithms and Evaluation  

```python
train, test = train_test_split(featureScaling, test_size = 0.30, random_state=0, shuffle=True)

print(train.shape)
print(test.shape)

X_test=test[["features1","features2","features3","features4","features5","features6","features7","features8","features9","features10","features11","features12","features13","features14","features15","features16","features17","features18","features19","features20","features21","features22","features23","features24","features25","features26","features27","features28","features29","features30"]]
y_test=test.diagnosis

X_train = train[["features1","features2","features3","features4","features5","features6","features7","features8","features9","features10","features11","features12","features13","features14","features15","features16","features17","features18","features19","features20","features21","features22","features23","features24","features25","features26","features27","features28","features29","features30"]]
y_train = train.diagnosis
```

# -> LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)

print(model)

expected = y_test
predicted = model.predict(X_test)

print(metrics.classification_report(expected, predicted))

print("confusion Matrix: " +'\n')

print(metrics.confusion_matrix(expected, predicted))
print('\n')
print("Training score: {:.2f}".format(model.score(X_train, y_train)))
print("Test score: {:.2f}".format(model.score(X_test, y_test)))
```

# -> KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier().fit(X_train, y_train)
print(model)
expected = y_test
predicted = model.predict(X_test)
print(metrics.classification_report(expected, predicted))
print("confusion Matrix: "+"\n")
print(metrics.confusion_matrix(expected, predicted))
print("\n")
print("Training score: {:.2f}".format(model.score(X_train, y_train)))
print("Test score: {:.2f}".format(model.score(X_test, y_test)))
```

# -> Cross Validation -> LogisticRegression

```python
features = dataset.drop("diagnosis", axis=1) 
labels = dataset["diagnosis"].copy()
```

```python
LogisticRegression = LogisticRegression(C=11)
scores = cross_val_score(LogisticRegression, features, labels, cv=10)
print("Cross-validation scores: {}".format(scores))
print("\n")
print("Average cross-validation score: {:.2f}".format(scores.mean()))
```

# -> Cross Validation -> KNeighborsClassifier

```python
KNeighborsClassifier = KNeighborsClassifier(n_neighbors=6)
scores = cross_val_score(KNeighborsClassifier, features, labels, cv=5)
print("Cross-validation scores: {}".format(scores))
print("\n")
print("Average cross-validation score: {:.2f}".format(scores.mean()))
```
